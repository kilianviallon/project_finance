```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
f = pd.read_csv('walmart_stock.csv')
f
```

```python
f.set_index('Date', inplace = True)
```

```python
ax = f['Open'].plot(label = 'Opening price',lw = 1,ls = '-.',c = 'r', alpha = 0.9)
ax.set_ylabel('prices')
plt.gcf().autofmt_xdate() # Auto fixes the overlap! (gfc = get current figure)
plt.legend()
plt.grid()
plt.show()
```
