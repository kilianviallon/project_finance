```python
def count_price(car):
    x = 0
    n = len(car)
    string = car.lower()
    for k in range(n):
        if string[k:k+5] == 'price' :
            x+=1
    return("there are {} price here".format(x))

count_price('Wow that is a nice pRice, very nice Price! I said price 3 times.')
```
