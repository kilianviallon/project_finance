```python
import matplotlib.pyplot as plt 
%matplotlib inline

import numpy as np

x = np.arange(0,100)
y = x*2
z = x**2

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,'b')
ax.set_title("title")
ax.set_xlabel('x')
ax.set_ylabel('y')



```

```python
fig = plt.figure(figsize = (10,10), dpi = 100)
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax = [ax1,ax2]
for a in ax :
    a.plot(x,x,'r')
    a.set_xlabel('x')
    a.set_ylabel('y')
plt.show()
```

```python
fig = plt.figure()
ax_z = fig.add_axes([0,0,1,1])
ax_y = fig.add_axes([0.2,0.5,.4,.4])
ax_z.plot(x,z,'b')
ax_z.set_xlabel('X')
ax_z.set_ylabel('Z')
ax_y.set_xlim([20,22])
ax_y.set_ylim([30,50])
ax_y.set_title('ZOOM')
ax_y.plot(x,y,'r')
ax_y.set_xlabel('X')
ax_y.set_ylabel('Y')
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
l = 0
for ax in axes :
    if l == 0:
        ax.plot(x,y,'b--')
        l+=1
    else :
        ax.plot(x,z,'r')
plt.show()
```
